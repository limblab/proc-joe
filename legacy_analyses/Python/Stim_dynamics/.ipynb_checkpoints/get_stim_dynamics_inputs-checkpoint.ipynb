```python
def generate_data(tgt_dir,output_data,go_cue_idx,hold_idx,max_len):
        trial_len = output_data.shape[1]
        inputs = np.zeros((output_data.shape[0],max_len,2)) # 1 input
        outputs = np.zeros((output_data.shape[0],max_len,output_data.shape[2])) # outputs are x- and y-vel

        for i_trial in range(output_data.shape[0]):
            inputs[i_trial,hold_idx:,0] = (tgt_dir[i_trial]) # determines target
            inputs[i_trial,go_cue_idx:,1] = 1 # serves as a go cue
            #inputs[i_trial,go_cue_idx:,2] = np.linspace(0,1,inputs.shape[1]-go_cue_idx)
            outputs[i_trial,go_cue_idx:go_cue_idx+trial_len,:] = output_data[i_trial,:,:]
            outputs[i_trial,go_cue_idx+trial_len:,:] = output_data[i_trial,-1,:]
        return (inputs,outputs)

def get_input_data(predict_emg,do_emg_norm,go_cue_idx,hold_idx):
    if(predict_emg):
        degs = np.array([0,90,180,270])
        deg_file_start = '/tf/GIT/proc-joe/Python/datasets/Han_20170203_emg_reach_'
        deg_file_end= 'deg_20ms.txt'
        output_data=np.array([])
        for d in range(len(degs)):
            deg_file = deg_file_start + str(degs[d]) + deg_file_end
            emg_temp = np.loadtxt(deg_file,delimiter=',')

            if(d == 0):
                output_data = np.zeros((len(degs),emg_temp.shape[0],emg_temp.shape[1]))
                tgt_dir = np.zeros((len(degs),))
            output_data[d,:,:] = np.expand_dims(emg_temp,axis=0)
            tgt_dir[d] = degs[d]*np.pi/180
        if(output_data.shape[1] > 75):
            output_data = output_data[:,:75,:]
        # normalize each emg signal by maximum signal
        if(do_emg_norm):
            max_mag = np.max(np.max(np.abs(output_data),axis=1),axis=0)
            max_mag = np.tile(np.expand_dims(np.expand_dims(max_mag,axis=0),axis=0),(output_data.shape[0],output_data.shape[1],1))
            output_data = np.multiply(output_data, 1/(max_mag))

    else:
        # generate desired trajectories for degs
        degs = np.array([0,45,90,135,180,225,270,315])*np.pi/180
        num_points = 75
        normal_dist = 5*norm.pdf(np.linspace(-2.5,2.5,num_points))

        for d  in range(len(degs)):
            vel_temp = np.transpose(normal_dist*np.array([[np.cos(degs[d])],[np.sin(degs[d])]]))
            if(d == 0):
                output_data = np.zeros((len(degs),num_points,2))
                tgt_dir = np.zeros((len(degs),))
            output_data[d,:,:] = np.expand_dims(vel_temp,axis=0)
            tgt_dir[d] = degs[d]


    # generate input and output data
    # input is a time vec starting at -1, then stepping to tgt_dir and staying there throughout the reach
    # output is x- and y-velocity at each time point. These are 0 until the step go cue, and then x_vel and y_vel measured during targets
    post_trial_idx = 25
    max_len = output_data.shape[1]+go_cue_idx+post_trial_idx


    x_train,y_train = generate_data(tgt_dir,output_data,go_cue_idx,hold_idx,max_len)
    x_test,y_test = x_train,y_train
    return x_train,y_train,x_test,y_test,degs
```
